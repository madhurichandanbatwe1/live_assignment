```python
cols=['SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
       'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
       'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']  
```

```python
for col in cols:
    plt.figure(figsize=(8, 6))
    ax = sns.countplot(x=col, data=df) #1=NO  #2=YES
    
    # Annotate the count plot
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')
    
    plt.title(f'Count of {col} is')
    plt.xlabel(f'{col}')
    plt.ylabel('Count')
    plt.show()  
```

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x='GENDER',data=df)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
age_labels = ['0-30','30-40', '41-50', '51-60', '61-70', '71-80']
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=6, labels=age_labels, right=False)
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='AGE_GROUP', hue='LUNG_CANCER', data=df)

# Annotate the count plot
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.title('Lung Cancer Cases by Age Group')
```

```python
for col in cols:
    plt.figure(figsize=(8,5))
    ax = sns.countplot(x=col, data=df,hue=df['LUNG_CANCER']) 
    
    # Annotate the count plot
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')
    
    plt.title(f'Lung Cancer cases by {col} group is')
    plt.xlabel(f'{col}')
    plt.ylabel('Count')
    plt.show()  
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot(x='GENDER', hue='LUNG_CANCER', data=df)

# Annotate the count plot
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

plt.title('Lung Cancer Cases by GENDER Group')
```
